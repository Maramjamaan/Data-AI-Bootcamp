### Q18 — Inheritance and Polymorphism

Create the following class hierarchy:

1. A **base class** `Shape` with:
   - A method `area()` that returns `0`.
   - A method `describe()` that returns `"I am a shape."`

2. A class `Rectangle` that **inherits** from `Shape`:
   - Has `width` and `height` attributes.
   - Overrides `area()` to return `width * height`.

3. A class `Circle` that **inherits** from `Shape`:
   - Has a `radius` attribute.
   - Overrides `area()` to return `3.14 * radius ** 2`.

**Test polymorphism:**
- Create a list containing a `Rectangle(5, 4)` and a `Circle(7)`.
- Loop through the list and print each shape's `area()`.

````python
# Q18:
class Shape :
  def area(self):
    return 0
  def describe(self):
    return f'I am Shape '

class Rectangle(Shape):
  def __init__ (self , height  ,width ):
    self.height  = height
    self.width = width

  def area(self):
    return self.height * self.width


class Circle(Shape) :
  def __init__ (self , radius ):
    self.radius = radius

  def area(self) :
    return 3.14 * self.radius ** 2


shapes_list = [Rectangle(5,4 ) ,Circle(7)]

for shape in shapes_list:
  shape_name = type(shape).__name__

  print(f'the area of {shape_name} is : {shape.area()}')
````

### B1 — Abstract Classes and Polymorphism

Using the `abc` module:

1. Create an **abstract class** `Vehicle` with:
   - An abstract method `start()`
   - An abstract method `stop()`

2. Create a class `ElectricCar` that inherits from `Vehicle`:
   - Has an attribute `brand`.
   - Implements `start()` → prints `"<brand> electric engine starting silently..."`
   - Implements `stop()` → prints `"<brand> has stopped. Battery saved."`

3. Create a class `GasCar` that inherits from `Vehicle`:
   - Has an attribute `brand`.
   - Implements `start()` → prints `"<brand> gas engine roaring to life!"`
   - Implements `stop()` → prints `"<brand> engine turned off."`

**Test:**
- Create a list with an `ElectricCar("Tesla")` and a `GasCar("Toyota")`.
- Loop through the list and call `start()` and `stop()` on each — demonstrating **polymorphism**.

````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass

class ElectricCar(Vehicle):
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        return f"{self.brand} electric engine starting silently..."

    def stop(self):
        return f"{self.brand} has stopped. Battery saved."

class GasCar(Vehicle):
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        return f"{self.brand} gas engine roaring to life!"

    def stop(self):
        return f"{self.brand} engine turned off."

cars = [ElectricCar("Tesla"), GasCar("Toyota")]


for car in cars:
    print(car.start())
    print(car.stop())
    print("-" * 40)
````
