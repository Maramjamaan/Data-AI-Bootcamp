## Q2) Encapsulation with BankAccount
Create a class `BankAccount` with a private attribute `__balance`, and methods `deposit(amount)` and `get_balance()`.
Try to modify the balance directly from outside and observe what happens.

```python
class BanckAccount:

  def __init__(self , balance, amount):
    self.__balance = balance
    self.amount = amount

  def get_balance(self) :
    return self.__balance + self.amount

  def deposit(self):
    return self.amount

acc1 = BanckAccount(300 ,8999)
print(f'deposit amount : {acc1.deposit()} \n account balance :{acc1.get_balance()}')
```

## Q4) Inheritance – Person & Student
Create class `Person` with `name` and method `introduce()`.
Create subclass `Student` that adds `age` and method `more_info()`.
Instantiate both and call their methods.

```python
class Person :

  def __init__(self, name):
    self.name= name
  def intreduce (self):
    return f'My name is:  {self.name}'

class Student(Person):

  def __init__ (self , name , age ) :
      super().__init__(name)
      self.age = age
  def more_info(self):
    return f'Student {self.name}  is {self.age} years old'

person = Person('Maram')
student= Student('Lamar',16)

print(person.intreduce())
print(student.more_info())
```

## Q5) Polymorphism – Animals
Create base class `Animal` with `speak()` returning "(silence)".
Create `Dog` and `Cat` that override `speak()`.
Iterate a mixed list and call `speak()` on each.

```python
class Animal :
    def speak(self):
      print('(silence)')


class Cat(Animal):
  def speak(self):
     print("Cat go's (Meow)")


class Dog (Animal):
  def speak(self):
    print("Dog go's (Woof)")

animals = [Animal(), Cat(), Dog()]
for a in animals:
    a.speak()
```

## Q6) Protected Attributes – Course
Create class `Course` with protected attributes `_name` and `_hours`, and method `display()`.
Create subclass `OnlineCourse` that updates these values and displays them.
Try accessing the protected attributes from outside.

```python
class Course :
  def __init__(self ,name , hours ):
    self._name = name
    self._hours = hours

  def display(self) :
    return f"Course: {self._name} , Hours: {self._hours}"

class OnlineCourse(Course) :
  def update(self, name, hours):
    self._name = name
    self._hours = hours

  def display(self) :
    return f"Online Course Name: {self._name},  Hours: {self._hours}"

course1 = OnlineCourse('python', 50)
print(course1.display())
course1.update('math 101', 60)
print(course1.display())
```

## Q8) Abstraction – Shapes
Use `abc.ABC` to create abstract class `Shape` with abstract method `area()`.
Create subclass `Rectangle(w, h)` that implements `area()` and test it.

```python
from abc import ABC , abstractmethod

class Shape (ABC):
  @abstractmethod
  def area(self):
    pass

class Rectangle(Shape):
  def __init__(self, width, height):
    self.width = width
    self.height = height

  def area(self):
    return self.width * self.height

rectangle = Rectangle(5, 10)
print(rectangle.area())
```

## Q9) Bicycle Ride
