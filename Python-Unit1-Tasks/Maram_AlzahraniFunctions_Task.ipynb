### Q1) Celsius → Fahrenheit
Implement `c_to_f(c)` that returns Fahrenheit using `F = C * 9/5 + 32`. Test with a few values.

````python
def c_to_f(c):
  F = (c * 9/5) + 32
  return F
temp1=c_to_f(24)
print(temp1)

temp2= c_to_f(55)
print(temp2)

temp3=c_to_f(10)
print(temp3)
````

### Q13) Default parameter + list-comprehension
Write `power_all(nums, exp=2)` that returns a new list with each number raised to `exp`.

````python
def power_all(nums, exp=2):

  return [n ** exp for n in nums]

lst = power_all([2, 4, 1, 3])
print(lst)

lst2 = power_all([2, 4, 1, 3], exp=3)
print(lst2)
````

#### Understanding Positional-only and Keyword-only Parameters
- **Positional-only** parameters must be passed **without** parameter names.
  - In Python syntax, these are declared **before** the `/` symbol in the function definition.
  - Example: `def func(a, b, /):` → both `a` and `b` must be given without names.
- **Keyword-only** parameters must be passed **with** parameter names.
  - In Python syntax, these are declared **after** the `*` symbol in the function definition.
  - Example: `def func(*, c, d):` → both `c` and `d` must be passed with names.

**Why use them?**
- Control how functions are called.
- Avoid accidental name clashes or improve readability.

**Quick demo:**
```python
def demo(a, b, /, *, sep='-'):
    return f"{a}{sep}{b}"

print(demo('hi', 'all', sep='*'))   #  works
demo(a='hi', b='all', sep='*')      #  TypeError (a and b are positional-only)
```


### Q15) Positional-only and keyword-only
Define `concat(a, b, /, *, sep='-')` that returns `"a{sep}b"`.
- `a` and `b` must be positional-only
- `sep` must be keyword-only

````python
def concat(a, b, /, *, sep='-'):
    return f"{a}{sep}{b}"
print(concat('hi', 'all', sep='*'))
````
