## **Lists Challenges & Solutions**

### Quick Review: Lists in Python
- **Ordered**: Items have a defined order.
- **Mutable**: You can change, add, or remove items.
- Can store mixed data types.
- Created using square brackets `[]`.
- Access elements using index, starting at 0.
- Common methods: `.append()`, `.remove()`, `.insert()`, `.sort()`.

**Lists Challenge 1:** Create a list of 5 cities and print the second city.

**Solution:**

```python
cities=['Riyadh','Makkah','Albaha','Jeddah','Hail']
print(cities[1])
```

**Lists Challenge 2:** Create a list of numbers from 1 to 10 using `range()` and print the list.

**Solution:**

```python
numlist=list(range(1,11))
print(numlist)
```

**Lists Challenge 3:** Given a list `nums = [5, 10, 15, 20]`, replace the second element with 99.

**Solution:**

```python
nums = [5, 10, 15, 20]
nums[1]=99
print(nums)
```

**Lists Challenge 4:** Append the word 'Python' to an empty list and print the list.

**Solution:**

```python
word=[]
word.append('Python')
print(word)
```

**Lists Challenge 5:** From the list `colors = ['red', 'green', 'blue', 'yellow']`, print only the first two elements.

**Solution:**

```python
colors = ['red', 'green', 'blue', 'yellow']
print(colors)
print(f'first two elements of the list is :{colors[:2]}')
```

## **Tuples Challenges & Solutions**

### Quick Review: Tuples in Python
- **Ordered**: Items have a defined order.
- **Immutable**: Cannot be changed after creation.
- Can store mixed data types.
- Created using parentheses `()`.
- Access elements using index, starting at 0.
- Useful when data should not change.

**Tuples Challenge 1:** Create a tuple with 4 fruits and print the last fruit.

**Solution:**

```python
fruits=['cherry','apple','banana','orange']
print(f' last item of the list is : {fruits[-1]}')
```

**Tuples Challenge 2:** Given a tuple `nums = (1, 2, 3)`, convert it to a list and add the number 4.

**Solution:**

```python
numstup = (1,2,3)
print(f'nums original tuple : {numstup}')

numlst= list(numstup)
numlst.append(4)
print(f'nums converted to a list and 4 is added ,{numlst}')

numstup=tuple(numlst)
print(f'nums tuple after adding 4 : {numstup}')
```

**Tuples Challenge 3:** Create a tuple of 3 countries and print its length.

**Solution:**

```python
contries = ('Saudi Arabia', 'Russia','China')
print(len(contries))
```

**Tuples Challenge 4:** Check if 'apple' exists in the tuple `fruits = ('apple', 'banana', 'cherry')`.

**Solution:**

```python
fruits2 = ('apple', 'banana', 'cherry')
print(fruits2)
print('Checks if apple is in the tuple: ','apple' in fruits2 )
```

**Tuples Challenge 5:** Create a tuple of your 3 favorite movies.

**Solution:**

```python
fav_movies = ('Sentimental Value','Eternity','Toy Story')
for f in fav_movies :
  print(f)
```

## **Dictionaries Challenges & Solutions**

### Quick Review: Dictionaries in Python
- Store data as **key-value pairs**.
- Keys must be unique and immutable.
- Values can be any data type.
- Created using curly braces `{}` with `key: value` pairs.
- Access values using keys.
- Common methods: `.keys()`, `.values()`, `.items()`.

**Dictionaries Challenge 1:** Create a dictionary for a student with keys: name, age, grade.

**Solution:**

```python
student = {'name': 'Areej' , 'age': 24 , 'grade': 98}
print(student)
```

**Dictionaries Challenge 2:** Add a key 'city' with value 'Riyadh' to the dictionary.

**Solution:**

```python
student = {'name': 'Areej' , 'age': 24 , 'grade': 98}
student['city']='Riyadh'
print(student)
```

**Dictionaries Challenge 3:** Remove the key 'grade' from the dictionary.

**Dictionaries Challenge 4:** Given `scores = {'Ali': 85, 'Sara': 90}`, print Sara's score.

**Solution:**

```python
scores = {'Ali': 85, 'Sara': 90}
print(f"Sara's score is : {scores['Sara']}")
```

**Dictionaries Challenge 5:** Loop through a dictionary and print all keys.

**Solution:**

```python
#Sense and Sensibility (1811)
fav_book = {'book':"Sense and Sensibility",
            'author': 'Jane Austin',
            'year published' : 1811,
            'publishsin oreder' : 'First novel'}

print(fav_book)

for book in fav_book :
  print('Dictionary Key : ',book)
```

## **Sets Challenges & Solutions**

### Quick Review: Sets in Python
- **Unordered**: No defined order.
- **Unique elements**: No duplicates allowed.
- Mutable: Can add or remove elements.
- Created using curly braces `{}` or `set()`.
- No indexing; use loops to access elements.
- Supports mathematical set operations like union and intersection.

**Sets Challenge 1:** Create a set with numbers 1, 2, 3 and print it.

**Solution:**

```python
set1={1,2,3}
print(set1)
print(type(set1))
```

**Sets Challenge 2:** Add number 4 to the set `{1, 2, 3}`.

**Solution:**

```python
set1={1,2,3}
set1.add(4)
print(set1)
```

**Sets Challenge 3:** Remove number 2 from the set `{1, 2, 3}`.

**Solution:**

```python
set2={1,2,3}
set2.remove(2)
print(set2)
```

**Sets Challenge 4:** Find the intersection of `{1, 2, 3}` and `{2, 3, 4}`.

**Solution:**

```python
setnum1={1,2,3}
setnum2 = {2,3,4}
print(setnum1 | setnum2)
```

**Sets Challenge 5:** Check if the number 5 exists in the set `{1, 2, 3, 4}`.

**Solution:**

```python
setA= { 1,2,3,4}
print(5 in setA)
```

## **Advanced Challenges & Solutions**
These challenges combine multiple data structures or simulate real-life scenarios.

**Lists Challenge (Advanced):** Create a list of ages and use list comprehension to get only ages above 18.

**Solution:**

```python
ages = [33,15,18,4,60,14,17,23,22,12,51,27]
adult=[age for age in ages if age>18 ]
print(adult)
```

**Tuples Challenge (Advanced):** Given a tuple of numbers, create a list containing only multiples of 3.

**Solution:**

```python
numbers = (5,6,7,8,1,2,3,4,6,3,16,21,34)
numberslist = list(numbers)
multiples3=[j for j in numberslist if j % 3 ==0  ]
print(multiples3)
```

**Dictionaries Challenge (Advanced):** Store employees in a list of dictionaries (name, department, salary) and print names in IT with salary > 8000.

**Solution:**

```python
employees = [{"name": "Ali", "department": "IT" , "salary": 16000},
             {"name": "Ahmed", "department": "HR" , "salary": 6000},
             {"name": "Sara", "department": "IT" , "salary": 10000},
             {"name": "Nada", "department": "PR" , "salary": 18000},
             {"name": "bayan", "department": "IT" , "salary": 6000}]

print("Employees in IT with salary > 8000:")
for employee in employees:
    if employee['department'] == 'IT' and employee['salary'] > 8000:
        print(employee['name'])
```

**Dictionaries Challenge (Advanced):** Given a dictionary of products and prices, print products sorted from highest to lowest price.

**Solution:**

```python
products = {'bag':1500 ,
            'shoe':430,
            'watch':700,
            'perfume':800,
            'iphone case':80,
            'f1 merch':400}

# Sort products by price from highest to lowest
sorted_product = sorted(products.items(), key=lambda product: product[1], reverse=True)
print(sorted_product)
```

**Sets Challenge (Advanced):** Check for duplicate phone numbers using a set.

**Solution:**

```python
phone_numbers = ['053323456','0593482535','053323456','0532385783','0593482535']
unique_phone_numbers = set(phone_numbers)

duplicates = []
for number in phone_numbers:
    if phone_numbers.count(number) > 1 and number not in duplicates:
        duplicates.append(number)

print("Duplicated numbers:", duplicates)
```

**Dictionaries Challenge (Advanced):** Count how many times each word appears in a sentence.

**Solution:**

```python
sentence= 'this taks is about is collections in pyhton collection task'
word= sentence.split()
wordcount= {}
for w in word:
  wordcount[w] = wordcount.get(w, 0) + 1
print(wordcount)
```

**Sets Challenge (Advanced):** Given two sets of students in different subjects, find students in both, in only one, and in either.

**Solution:**

```python
Math101 = {'lamr', 'joud', 'rana','nada','maram' , 'leena'}
El112 = {'maram', 'leena', 'sara','raghad', 'nada'}

both = Math101 & El112
print(both)

one= Math101.symmetric_difference(El112)
print(one)

either=Math101.union(El112)
print(either)
```
