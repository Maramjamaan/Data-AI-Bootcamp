###15.Write a Python program that asks the user to enter their **age** and whether they are a **student** (`yes` or `no`).

Then, use **logical operators (`and`, `or`, `not`)** to check if they can get a **discount**.

**Rules:**
1. If the person is **under 18** **or** is a **student**, they get a discount.
2. Otherwise, no discount.

````python
age = int(input("Enter your age: "))
student = input("Are you a student? (yes/no): ")

is_student = (student == 'yes')

# Rule: If the person is under 18 OR is a student, they get a discount.
if age < 18 or is_student:
    print("You get a discount!")
else:
    print("No discount ")
````
