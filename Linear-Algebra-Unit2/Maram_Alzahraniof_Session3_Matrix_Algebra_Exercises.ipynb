```python
import numpy as np
```

```python
A = np.array([[1, 3],
              [2, 4]])

B = np.array([[2, 0],
              [1, 2]])
```

*   Compute A + B
* Compute A @ B
* Compute B @ A
* Are A @ B and B @ A equal?




```python
print(f'A+B =\n {A+B}')
print()
print(f'A@B =\n {A@B}')
print()
print(f'B@A =\n {B@A}')
print()
print('A @ B dont equal B @ A ')
```

#Determinant & Inverse

1.   Compute determinant of A
2. Compute inverse of A
3. Verify A @ A⁻¹ = I




```python
det_A = np.linalg.det(A)
print(f'Determinant of A = {det_A}')
print()

inv_A= np.linalg.inv(A)
print(f"inversive of A :\n {inv_A}")
print()

print(f' A @ A⁻¹ = I  \n {A@inv_A}' )
```

#Solve a System

Solve:

* 3x + 2y = 16
* 4x − y = 9

Tasks:

1. Write matrix A and vector b

2. Use np.linalg.solve

3. Verify by substitution

```python
A= np.array([[3,4],[2,-1]])

b= np.array([16,9])

eq =np.linalg.solve(A,b)

print(f'x = {eq[0]}')
print(f'y = {eq[1]}')

print(f'3x + 2y = {3*eq[0] + 2*eq[1]}')
print(f'4x - y = {4*eq[0] - eq[1]}')
```
