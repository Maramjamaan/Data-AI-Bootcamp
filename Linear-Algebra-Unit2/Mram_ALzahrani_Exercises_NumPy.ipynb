#Exercise 5 – Matrix Multiplication

Given:
```
A = np.array([[1, 2],
              [3, 4]])


B = np.array([[2, 0],
              [1, 2]])
```

* Compute element-wise multiplication.
* Compute matrix multiplication.
* Explain the difference in a comment.

````python
A = np.array([[1, 2],
              [3, 4]])


B = np.array([[2, 0],
              [1, 2]])

element_wise= A * B

print(element_wise)

#element wise matches the input dimension exactly


matrix_multiplication= A @ B

print("")
print(matrix_multiplication)

#Matrix multiplication outer dimension of inputs
````

#Exercise 6 – Statistics Practice

Create a dataset of exam scores:

```
scores = np.array([
    [70, 80, 90],
    [85, 75, 95],
    [60, 65, 70],
    [90, 88, 92]
])

```

* Compute mean score per student.

* Compute mean score per subject.

* Find the highest and lowest score.

* Compute standard deviation.

````python
scores = np.array([
    [70, 80, 90],
    [85, 75, 95],
    [60, 65, 70],
    [90, 88, 92]
])

print(np.mean(scores))
print(np.mean(scores,axis=1))
print(np.max(scores))
print(np.min(scores))
print(np.std(scores))
````
