### Difference Between (n,) and (n,1)

-(n,) represents a one-dimensional vector with a single axis.

-(n,1) represents a two-dimensional column vector with one column.

```python
# Apply element-wise operations

scores = np.array([
                  [80, 90, 85],
                  [70, 75, 78],
                  [95, 88, 92],
                  [60, 65, 70]])

extra_credit = np.array([[2, 1, 3],
                         [1, 2, 1],
                         [3, 2, 2],
                         [1, 1, 1]])

print(f'Scores Data : \n{scores}')
print('-'*40)
print(f'Extra Credit : \n{extra_credit}')
print('-'*40)
print(f'Addition :\n {scores + extra_credit}')
print('-'*40)
print(f'Subtraction :\n {scores - extra_credit}')
print('-'*40)
print(f'Multiplication :\n {scores * extra_credit}')
print('-'*40)
print(f'Division :\n {scores / extra_credit}')
```

```python
# Apply aggregations:
print(f'Sum of Scores : {scores.sum()}')
print(f'Mean of Scores : {scores.mean()}')
print(f'Max of Scores : {scores.max()}')
print(f'Min of Scores : {scores.min()}')
print(f'Standard Deviation of Scores : {scores.std()}')

print('-'*40)

print(f'Sum of Extra Credit : {extra_credit.sum()}')
print(f'Mean of Extra Credit : {extra_credit.mean()}')
print(f'Max of Extra Credit : {extra_credit.max()}')
print(f'Min of Extra Credit : {extra_credit.min()}')
print(f'Standard Deviation of Extra Credit : {extra_credit.std()}')
```

```python
# 1. Create two vectors v and w.
v= np.array([12.50, 11.99, 13.00])
w=np.array([45.00, 42.50, 46.99])
print(f'victor v : {v}')
print(f'victor w : {w}')
print('-'*40)
# Vector addition
print(f'Victor Addition : {v+w}' )
print('-'*40)
# Vector subtraction
print(f'Victor Subtraction : {v-w}' )
print('-'*40)
# Scalar multiplication
scalar1=np.array(5)
print(f'Scalar multiplication :{v*scalar1} ')
print('-'*40)
# Dot product
print(f'Dot product : {np.dot(v,w)}')
print('-'*40)
# Norm
norm_v = np.linalg.norm(v)
norm_w = np.linalg.norm(w)
print("||v|| =", norm_v)
print()
print("||w|| =", norm_w)
```

```python
# 5.Create two vectors and test
v1 = np.array([1,2])
v2 = np.array([3,4])
print(f'v1 : {v1} \nv2 : {v2}')
print()
A= np.column_stack([v1,v2])
print(f"Rank([v1, v2]) = {np.linalg.matrix_rank(A)} ")
print()

# Are they linearly independent?
print('The vectors are linearly independent.')

# Do they form a basis for R2? yes the rank is 2
print("The vectors form a basis for R².")
```

```python
# Create a linear combination:
# a * v + b * w
a=2
b=3

lin_comp = a*v1 + b*v2
print(f'linear combination : {lin_comp}')
```

```python
# Apply a linear transformation using a matrix:
# Scaling
scale_mat = np.array([[2,4],[6,3]])
scale_vec = scale_mat.dot(v1)
print(f'Original matrix : \n{scale_mat}')
print()
print(f'Scaled : {scale_vec}')
print()
# o Rotation
rotation = np.array([[0,-1],
                      [1,0]])
rotated_vector = rotation @ v1

print(f'Original matrix : \n{rotation}')
print()
print("Rotated Vector:", rotated_vector)
```
