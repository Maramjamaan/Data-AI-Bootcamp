#Exercise 1 — Vector basics & shapes

## Task

Create vectors:

v = [2, -1, 3]

w = [1, 4, -2]

* Print:

* shape, ndim

* Convert v into a column vector shape (3,1)

```python
import numpy as np
v = np.array([2,-1,3])
w= np.array([1,4,-1])

print(f'V vector : {v} , Shape : {v.shape} , Dimention : {v.ndim}')
print(f'W vector : {w} , Shape : {w.shape} , Dimention : {w.ndim} ')

V_col =  v.reshape(-1, 1)
print(f'V vector converted into column vector  : {V_col.shape}')
```

# Exercise 2 — Norm & normalization

## Task

* Compute ||v||

* Create a normalized vector v_unit = v / ||v||



```python
norm_v = np.linalg.norm(v)
print("||v|| =", norm_v)

v_uni = v / norm_v
print(f'normalized vector : {v_uni} ' )
```
